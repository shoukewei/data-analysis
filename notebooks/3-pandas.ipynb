## Filtering and Transformation

### Selecting columns

```python
# Single column → Series
alcohol = df["alcohol"]

# Multiple columns → DataFrame
subset = df[["alcohol", "pH", "quality"]]
print(subset.shape)   # (1599, 3)
```

### Label-based and position-based indexing

```python
# .loc: rows 0–4, named columns
print(df.loc[0:4, ["alcohol", "quality"]])

# .iloc: first 5 rows, last 2 columns
print(df.iloc[:5, -2:])
```

### Boolean filtering

```python
# Wines with quality >= 7
high_quality = df[df["quality"] >= 7]
print(high_quality.shape)   # (217, 12)

# High quality AND low volatile acidity
mask = (df["quality"] >= 7) & (df["volatile acidity"] < 0.4)
print(df[mask].shape)       # (125, 12)

# Wines with alcohol outside the 10–12% range
outliers = df[~df["alcohol"].between(10, 12)]
print(outliers.shape)       # (821, 12)
```

### `.query()` syntax

```python
# Equivalent to df[(df["quality"] >= 7) & (df["alcohol"] > 11)]
result = df.query("quality >= 7 and alcohol > 11")
print(result.shape)   # (141, 12)
```

### Adding and transforming columns

```python
# Acidity ratio: fixed / volatile
df["acidity_ratio"] = df["fixed acidity"] / df["volatile acidity"]

# Binary quality label: 1 = good (quality >= 7), 0 = average
df["quality_label"] = (df["quality"] >= 7).astype(int)

# Log-transform residual sugar to reduce right skew
import numpy as np
df["log_sugar"] = np.log1p(df["residual sugar"])

print(df[["acidity_ratio", "quality_label", "log_sugar"]].head(3))
```

### Applying functions with `.apply()`

```python
# Classify alcohol content into tiers
def alcohol_tier(x):
    if x < 9.5:
        return "low"
    elif x < 11.5:
        return "medium"
    else:
        return "high"

df["alcohol_tier"] = df["alcohol"].apply(alcohol_tier)
print(df["alcohol_tier"].value_counts())
```

### Renaming and dropping columns

```python
# Rename columns for cleaner names
df = df.rename(columns={
    "fixed acidity":        "fixed_acidity",
    "volatile acidity":     "volatile_acidity",
    "citric acid":          "citric_acid",
    "residual sugar":       "residual_sugar",
    "free sulfur dioxide":  "free_so2",
    "total sulfur dioxide": "total_so2",
})

# Drop engineered helper columns
df = df.drop(columns=["log_sugar", "acidity_ratio"])

print(df.columns.tolist())
```

## Aggregation and Group Operations

### Descriptive statistics on a DataFrame

```python
print(df["alcohol"].mean())     # 10.42
print(df["alcohol"].median())   # 10.20
print(df["alcohol"].std())      #  1.07
print(df["alcohol"].quantile([0.25, 0.75]))
```

```python
corr = df[["fixed_acidity", "volatile_acidity", "alcohol", "quality"]].corr()
print(corr.round(2))
```

### `groupby` operations

```python
# Mean of every column, grouped by quality score
quality_groups = df.groupby("quality").mean(numeric_only=True).round(2)
print(quality_groups[["alcohol", "volatile_acidity", "pH"]])
```

### Multiple aggregations with `.agg()`

```python
stats = df.groupby("quality")["alcohol"].agg(["mean", "std", "min", "max", "count"])
print(stats.round(2))
```

```python
agg_dict = {
    "alcohol":          ["mean", "std"],
    "volatile_acidity": ["mean", "min", "max"],
    "quality":          "count",
}
result = df.groupby("alcohol_tier").agg(agg_dict).round(2)
print(result)
```

### Pivot tables

```python
pivot = pd.pivot_table(
    df,
    values="alcohol",
    index="quality",
    columns="alcohol_tier",
    aggfunc="mean",
).round(2)
print(pivot)
```

### Value counts and frequency tables

```python
print(df["quality"].value_counts(normalize=True).sort_index().round(3))
```

## Working with Time-Indexed Data

### Building a DatetimeIndex

```python
import pandas as pd
import numpy as np

url = (
    "https://raw.githubusercontent.com/mlflow/mlflow/"
    "master/tests/datasets/winequality-red.csv"
)
df = pd.read_csv(url, sep=";")
df.columns = df.columns.str.replace(" ", "_")

# Create a daily DatetimeIndex with 1,599 business days
dates = pd.date_range(start="2020-01-01", periods=len(df), freq="D")
df.index = dates
df.index.name = "date"

print(df.index[:5])
```

### Slicing by date

```python
# A single month
jan_2020 = df.loc["2020-01"]
print(jan_2020.shape)   # (31, 12)

# A date range
q1_2020 = df.loc["2020-01-01":"2020-03-31"]
print(q1_2020.shape)    # (91, 12)

# A full year
year_2020 = df.loc["2020"]
print(year_2020.shape)  # (366, 12)
```

### Resampling

```python
monthly = df[["alcohol", "quality"]].resample("ME").mean().round(2)
print(monthly.head(6))
```

### Rolling windows

```python
# 30-day rolling mean of alcohol content
df["alcohol_30d_mean"] = df["alcohol"].rolling(window=30).mean()

# 30-day rolling standard deviation — a measure of variability
df["alcohol_30d_std"] = df["alcohol"].rolling(window=30).std()

print(df[["alcohol", "alcohol_30d_mean", "alcohol_30d_std"]].iloc[28:33].round(3))
```

### Expanding windows and cumulative statistics

```python
# Cumulative mean of quality over time
df["quality_cumulative_mean"] = df["quality"].expanding().mean()

print(df[["quality", "quality_cumulative_mean"]].head(5).round(3))
```

### Shifting and lagged features

```python
# Lag-1 alcohol: previous day's alcohol reading
df["alcohol_lag1"] = df["alcohol"].shift(1)

# Day-over-day change in alcohol
df["alcohol_diff"] = df["alcohol"].diff(1)

print(df[["alcohol", "alcohol_lag1", "alcohol_diff"]].head(4).round(3))
```
