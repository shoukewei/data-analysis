```python
import polars as pl
import glob
import os

# Columns you actually need (keeps file size small and fast)
cols = ["FlightDate", "Reporting_Airline", "Origin", "Dest",
        "DepDelay", "ArrDelay", "AirTime", "Distance",
        "Cancelled", "CarrierDelay", "WeatherDelay"]

csv_pattern = "../data/flights/On_Time_Reporting_Carrier_On_Time_Performance_(1987_present)_2023_*.csv"

print("Starting CSV → Parquet conversion...\n")

for csv_path in sorted(glob.glob(csv_pattern)):
    parquet_path = csv_path.replace(".csv", ".parquet")
    
    if os.path.exists(parquet_path):
        print(f"✅ Already exists: {os.path.basename(parquet_path)}")
        continue
    
    print(f"Converting {os.path.basename(csv_path)} ...")
    
    df = pl.read_csv(
        csv_path,
        columns=cols,
        infer_schema_length=10000,
        schema_overrides={
            "DepDelay": pl.Float64,
            "ArrDelay": pl.Float64,
            "AirTime": pl.Float64,
            "Distance": pl.Float64
        }
    )
    
    df.write_parquet(parquet_path, compression="snappy")
    
    size_mb = os.path.getsize(parquet_path) / (1024 * 1024)
    print(f"   → Saved {os.path.basename(parquet_path)} ({size_mb:.1f} MB)\n")

print("All conversions finished!")
```
